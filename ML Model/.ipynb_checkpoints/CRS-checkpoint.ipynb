```python
import pandas as pd
```

```python
df= pd.read_csv('Crop_recommendation.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
x = df.drop('label', axis = 'columns')
x.head()
```

```python
y = df.label
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, random_state = 1, test_size = 0.2)
```

```python
X_train.info()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred1 = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
logistic_reg_acc = accuracy_score(y_test, y_pred1)
print('The Accuracy of Logistic Model is ' + str((logistic_reg_acc)*100) + '%')
```

```python
from sklearn.tree import DecisionTreeClassifier
model_2 = DecisionTreeClassifier()
model_2.fit(X_train, y_train)
y_pred2 = model_2.predict(X_test)
```

```python
decision_tree_acc = accuracy_score(y_test, y_pred2)
print('The Accuracy of Decision Tree Model is ' + str((decision_tree_acc)*100) + '%')
```

```python
from sklearn.ensemble import RandomForestClassifier
model_3 = RandomForestClassifier()
model_3.fit(X_train, y_train)
y_pred3 = model_3.predict(X_test)
```

```python
random_forest_acc = accuracy_score(y_test, y_pred3)
print('The Accuracy of Random Forest Model is ' + str((random_forest_acc)*100) + '%')
```
